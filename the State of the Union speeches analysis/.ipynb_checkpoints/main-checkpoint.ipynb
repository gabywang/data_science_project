## Abstract

While analyzing the State of the Union speeches across decades and decades, we asked questions on how much the President's speeches would change over time. For example, we wanted to see if Presidents would become more concise in their language. Through the trend in stop words and sentence length, we saw that Presidents tend to be more brief in their speeches, which supports the idea of conciseness within their speeches as the years went by. Being concise is a very important trait within a speech because the ideas are clear and straightforward. Ideas usually get lost within wordy speeches, which thereby nullifies the informational aspect of the speech. Also, we wanted to see how similar the speeches between Presidents were, and concluded that Presidents living in similar eras had similar speeches. Our own exploratory analysis wanted to further the topic on consciseness within the speeches by comparing the different political parties. Today's political climate is so polarized, so it was interesting to explore which party had Presidents that gave more concise speeches. We ultimately see that Republicans had more concise speeches than the Democrats. 


## Introduction

The origin of the State of the Union Address is derived from the U.S. Constitution, stating that the President has a respnsibility in giving a informational report to Congress concerning the current state of the nation and also outlines the President's legislative agenda for what he wants to accomplish. Since President Wilson's address in 1913, presidents have been giving the speeches in person to garner more support for their legislative agendas. Through outlets such as radio, television, and the Internet, a rise in technological advances has allowed Presidents to speak directly to the American people.
